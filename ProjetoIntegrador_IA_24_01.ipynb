```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import MaxAbsScaler
import pandas as pd

review = pd.read_csv('/content/drive/MyDrive/DataSets/imdb-reviews-pt-br.csv')
review.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(review['text_pt'], review['sentiment'], test_size=0.2, random_state=42)

vectorizer = TfidfVectorizer(max_features=10000)
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)
#print('X_test; ', X_test_vectorized)
#print('X_train; ', X_train_vectorized)
```

```python
scaler = MaxAbsScaler()
X_train_normalized = scaler.fit_transform(X_train_vectorized)
X_test_normalized = scaler.transform(X_test_vectorized)
```

```python
classifier = LinearSVC()
classifier.fit(X_train_normalized, y_train)

y_predict = classifier.predict(X_test_normalized)
print('Acurácia: ', accuracy_score(y_test, y_predict))
print(classification_report(y_test, y_predict))
```

```python
new_text = ["O filme foi fantástico!", "Livro horrível."]
new_text_vectorized = vectorizer.transform(new_text)
new_text_normalized = scaler.transform(new_text_vectorized)
predictions = classifier.predict(new_text_normalized)
print("Predições para os novos textos:", predictions)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB  # Importar o classificador Naive Bayes
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import MaxAbsScaler
import pandas as pd

# Carregar os dados
review = pd.read_csv('/content/drive/MyDrive/DataSets/imdb-reviews-pt-br.csv')

# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(review['text_pt'], review['sentiment'], test_size=0.2, random_state=42)

# Vetorização dos textos
vectorizer = TfidfVectorizer(max_features=10000)
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

# Normalização dos dados
scaler = MaxAbsScaler()
X_train_normalized = scaler.fit_transform(X_train_vectorized)
X_test_normalized = scaler.transform(X_test_vectorized)

# Classificação
classifier = MultinomialNB()
classifier.fit(X_train_normalized, y_train)

# Previsões
y_predict = classifier.predict(X_test_normalized)

# Avaliação do modelo
print('Acurácia: ', accuracy_score(y_test, y_predict))
print(classification_report(y_test, y_predict))

# Exemplo de novos textos para prever
new_text = ["O filme foi fantástico!", "Livro horrível."]
new_text_vectorized = vectorizer.transform(new_text)
new_text_normalized = scaler.transform(new_text_vectorized)
predictions = classifier.predict(new_text_normalized)
print("Predições para os novos textos:", predictions)
```
